```python
import random
game_list = ['stone', 'paper', 'scissors']
games_number = 7
games_to_win = 3

# Human player picks
def choice():
    while True:
        try:
            x = input('Please enter stone, paper or scissors: ')
        except ValueError:
            print("Sorry, I didn't understand that.")
            continue
        if (x == 'paper') or ( x == 'stone') or (x == 'scissors'):
            return x
            break
        else:
            print('Please enter again')

# Machine random choice
def game(list):
    return random.choice(list)

# Human vs Machine
def combat():
    x = game(game_list) # Macnine's pick
    y = choice() # Human's pick
    if (x == 'stone') and (y == 'paper'):
        print("Machine's pick: ", x)
        return 2
    elif (x == 'stone') and (y == 'scissors'):
        print("Machine's pick: ", x)
        return 1
    elif (x == 'stone') and (y == 'stone'):
        print("Machine's pick: ", x)
        return 0
    elif (x == 'paper') and (y == 'paper'):
        print("Machine's pick: ", x)
        return 0
    elif (x == 'paper') and (y == 'scissors'):
        print("Machine's pick: ", x)
        return 2
    elif (x == 'paper') and (y == 'stone'):
        print("Machine's pick: ", x)
        return 1
    elif (x == 'scissors') and (y == 'paper'):
        print("Machine's pick: ", x)
        return 1
    elif (x == 'scissors') and (y == 'scissors'):
        print("Machine's pick: ", x)
        return 0
    elif (x == 'scissors') and (y == 'stone'):
        print("Machine's pick: ", x)
        return 2
    
def score():
    x = 0 # Machine win
    y = 0 # Human win
    for i in range(games_number):
        z = combat()
        if z == 1:
            x += 1
            print('Current score Machine vs Human ', x,':',y)
        elif z == 2:
            y += 1
            print('Current score Machine vs Human ', x,':',y)
        elif z == 0:
            print('Current score Machine vs Human ', x,':',y)
        if x == games_to_win:
            print ('Machine win!')
            break
        elif y == games_to_win:
            print ('Human win!')
            break
score()
```
